```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('datacleaning for holiday.csv')
```

```python
df = df.drop(columns = ['Unnamed: 0'])
```

```python
df
```

```python
dates = []

for row in df['Date']:
    dates.append(datetime.strftime(datetime.strptime(row,'%d/%m/%Y'),'%Y%m%d'))
    
df['Date'] = dates
```

```python
df.insert(0, 'TYPE', 'vevent')
```

```python
df.to_csv('cleanCalendar.csv')
```
