```python
import pandas as pd

df = pd.read_json('australian_postcodes.json')
```

```python
df = df[['id','postcode','locality','state','long','lat']]
df
```

```python
df[df.duplicated(subset=['postcode','locality', 'state'], keep=False)]
```

```python
df = df.drop_duplicates(subset=['postcode','locality', 'state'], keep='first', ignore_index = True)
```

```python
for i,val in enumerate(df.values):
    df['locality'][i] = df['locality'][i].title()
```

```python
df
```

```python
df.to_csv('postcodes.csv', index = False)
```
