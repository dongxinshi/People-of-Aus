```python
import pandas as pd
import json
df = pd.read_json('worshipWaysRelations.geojson')
```

```python

df['features'][0]
```

```python

df_clean = pd.DataFrame(list(df['features']))
df_clean
```

```python
df1 = pd.DataFrame(list(df_clean['properties']))[['amenity','denomination','name','religion']]
df1
```

```python
df2= pd.DataFrame(list(df_clean['geometry']))['coordinates']
df3 = pd.DataFrame()

```

```python
for i,row in enumerate(df2):
    df2[i] = df2[i][0]
```

```python
for row in df2:
    if len(row)==2:
        print(row)
```

```python
lonlist = []
latlist = []



for row in df2:
    if(len(row) != 2):
        
        lonsum = 0
        latsum = 0
        i = 0

        for val in row:

            lonsum += val[0]
            latsum += val[1]
            i+=1

        lonlist.append(lonsum/i)
        latlist.append(latsum/i)
        
    else:
        lonlist.append(row[0])
        latlist.append(row[1])
    

df3['lon'] = lonlist
df3['lat'] = latlist

df3
```

```python
df4 = pd.DataFrame(df_clean['id'])

for i,row in enumerate(df4['id']):
    df4['id'][i] = row.strip('way/')

df4
#df4['id'].strip('node/')
#.strip('node/')
```

```python
df1['id'] = df4['id']
df1[['lon','lat']] = df3[['lon','lat']]
```

```python
df1
```

```python
df1 = df1.dropna(subset=['amenity', 'name','religion','id','lon','lat']).reset_index(drop=True)
df1
```

```python
for i,row in enumerate(df1['denomination']):
    
    if (type(row) == float):
        df1['denomination'][i] = 'Unknown'
```

```python
import re

cleanvals = []

for i,row in enumerate(df1['name']):
    
    if(row.isascii() == False):
        print(row)
        if (row == 'Chùa Bồ Đề'):
            cleanvals.append('Chua Bo De')
        else:

            cleanvals.append(re.sub(r'[^\x00-\x7F]+',"'", row))
        
    else:
        cleanvals.append(row)
        
```

```python
df1['name'] = cleanvals
```

```python
for i,row in enumerate(df1['name']):
    if(row.isascii() == False):
        print(row)
```

```python
for i,row in enumerate(df1['denomination']):
    df1['denomination'][i] = df1['denomination'][i].replace('_',' ').title()
    df1['religion'][i] = df1['religion'][i].replace('_',' ').title()
    df1['name'][i] = df1['name'][i].replace('_',' ').title()
```
