```python
import pandas as pd
import json
df = pd.read_json('communityWays.geojson')
```

```python
df
```

```python

df_clean = pd.DataFrame(list(df['features']))
df_clean['properties'][1].keys()
```

```python
df1 = pd.DataFrame(list(df_clean['properties']))[['amenity','name']]
df1
```

```python
df2= pd.DataFrame(list(df_clean['geometry']))['coordinates']
df3 = pd.DataFrame()

for i,row in enumerate(df2):
    df2[i] = df2[i][0]
```

```python
lonlist = []
latlist = []



for row in df2:
    if(len(row) != 2):
        
        lonsum = 0
        latsum = 0
        i = 0

        for val in row:

            lonsum += val[0]
            latsum += val[1]
            i+=1

        lonlist.append(lonsum/i)
        latlist.append(latsum/i)
        
    else:
        lonlist.append(row[0])
        latlist.append(row[1])
    

df3['lon'] = lonlist
df3['lat'] = latlist

df3
```

```python
df2
```

```python
df4 = pd.DataFrame(df_clean['id'])

for i,row in enumerate(df4['id']):
    df4['id'][i] = row.strip('way/')

df4
#df4['id'].strip('node/')
#.strip('node/')
```

```python
df1['id'] = df4['id']
df1[['lon','lat']] = df3[['lon','lat']]
```

```python
df1 = df1.dropna().reset_index(drop=True)
df1
```

```python
import re

cleanvals = []

for i,row in enumerate(df1['name']):
    
    if(row.isascii() == False):
        if (row == 'recréation center'):
            cleanvals.append('recreation center')
        else:
            #df1['name'][i] = re.sub(r'[^\x00-\x7F]+',"'", row)
            cleanvals.append(re.sub(r'[^\x00-\x7F]+',"'", row))
            #print(row)
            #print(re.sub(r'[^\x00-\x7F]+',"'", row))
        
    else:
        cleanvals.append(row)
        
```

```python
df1['name'] = cleanvals
```

```python
for i,row in enumerate(df1['name']):
    if(row.isascii() == False):
        print(row)
```

```python
for i,row in enumerate(df1['name']):
    df1['name'][i] = df1['name'][i].title()
```
