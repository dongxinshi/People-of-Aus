```python
import pandas as pd

df = pd.read_json('australian_postcodes.json')
```

```python
df
```

```python
df = df[['id','postcode','locality','state','long','lat']]
df
```

```python
df[df['postcode'] == 820]
```

```python
for i,val in enumerate(df.values):
    df['locality'][i] = df['locality'][i].title()
```

```python
df
```

```python
post_list = dict()

for i,row in enumerate(df['postcode']):
    if row in post_list.keys():
        post_list[row].append([df['long'][i],df['lat'][i]])
    else:
        post_list[row] = [[df['long'][i],df['lat'][i]]]
```

```python
lonlist = []
latlist = []

for key in post_list:

        lonsum = 0
        latsum = 0
        i = 0

        for val in post_list[key]:
            lonsum += val[0]
            latsum += val[1]
            i+=1

        lonlist.append(lonsum/i)
        latlist.append(latsum/i)
        


```

```python
ting = {'postcode': list(post_list.keys()),'lon':lonlist,'lat':latlist}
```

```python
output = pd.DataFrame(ting)
```

```python
output
```

```python
output.to_csv('postcodes.csv', index = False)
```
