```python
import pandas as pd

df = pd.read_excel('school-profile-2008-2021d13a2f404c94637ead88ff00003e0139.xlsx', sheet_name = 1)
```

```python
df = df[['ACARA SML ID','Calendar Year','ICSEA Percentile','Language Background Other Than English - Yes (%)','Language Background Other Than English - No (%)','Language Background Other Than English - Not Stated (%)','Total Enrolments']]
```

```python
df
```

```python
df.isna().sum()
```

```python
na_list = []

for val in df['ACARA SML ID'].unique():
    if (df[df['ACARA SML ID'] == val].isna().sum().sum()>0):
        na_list.append(val)
```

```python
df = df.drop(df[df['ACARA SML ID'].isin(na_list)].index)
```

```python
clean_cols = []
for col in df.columns:
    clean_cols.append(col.replace(' ','_').lower().replace('_(%)','').replace('-','_').replace('___','_'))
    
df.columns = clean_cols
```
