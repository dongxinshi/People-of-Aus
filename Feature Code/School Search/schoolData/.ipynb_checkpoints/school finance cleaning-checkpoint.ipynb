```python
import pandas as pd
```

```python
df = pd.read_excel('privateSchoolFinanceData.xlsx')
```

```python
df
```

```python
df.columns = ['school_name','suburb','cost','cost','cost','cost1']
```

```python

df = df.groupby(df.columns, axis=1).sum()
```

```python
cost_list = []

for i,row in enumerate(df.values):
    if(df.cost[i] == 0):
        cost_list.append(df.cost[i]/3)
    else:
        cost_list.append((df.cost[i] + float(str(df.cost1[i]).strip('$').replace(',','')))/4)
```

```python
df['school_cost'] = cost_list
```

```python
df = df[['school_name','suburb','school_cost']]
```

```python
df[df['school_name'] == "Ruyton Girls' School"]
```

```python
df = df.drop(df[df['school_cost'] <100].index)
```

```python
df = df.groupby(['school_name','suburb']).mean().reset_index()
```

```python
df['suburb'] = df['suburb'].apply(lambda x: x.title())
```

```python
df
```

```python
df1 = pd.read_csv('clean_school_data.csv')

df2 = pd.read_csv('suburb_data.csv')
```

```python
df1 = df1.join(df2[['acara_sml_id','suburb']].set_index('acara_sml_id'),on = 'acara_sml_id')
```

```python
id_list = []

for i,row in enumerate(df.values):
    id_list.append((df1[(df1['school_name'].apply(lambda x: x.lower()) == df['school_name'][i].lower()) & (df1['suburb'].apply(lambda x: x.lower()) == df['suburb'][i].lower())]['acara_sml_id']))
    
```

```python
df['acara_sml_id'] = id_list
```

```python
df['acara_sml_id'] = df['acara_sml_id'].apply(lambda x: x.to_string().split(' ')[-1])
```

```python
df = df.drop(df[df['acara_sml_id'] == ')'].index)
```

```python
df = df.drop(columns = ['school_name','suburb'])
```
