```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_excel('school-location-2021e23a2f404c94637ead88ff00003e0139.xlsx', sheet_name=1)

df1 = pd.read_excel('school-profile-2021c03a2f404c94637ead88ff00003e0139.xlsx', sheet_name=1)
```

```python
df = df[['Latitude','Longitude','Local Government Area', 'ACARA SML ID']]
```

```python
df = df.rename(columns = {'Local Government Area Name':'LGA Name',
                    'Local Government Area': 'LGA Code'})
```

```python

df1 = df1.drop(columns = ['Location AGE ID','School AGE ID','Campus Type','Rolled Reporting Description','Governing Body','Governing Body URL','Geolocation'])
```

```python
df1.describe(include='all')
```

```python
df1.columns
```

```python
df = df[df['ACARA SML ID'].isin(list(df1['ACARA SML ID'].values))].reset_index()
```

```python
df = df1.join(df.set_index('ACARA SML ID'),on='ACARA SML ID')
```

```python
df
```

```python
df['School URL'] = df['School URL'].apply(lambda x: 'Unknown' if type(x) == float else x)
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset = ['ICSEA','Teaching Staff','index']).reset_index()
```

```python
df.isna().sum()
```

```python
clean_cols = []

for col in df.columns:
    clean_cols.append(col.replace(' ','_').lower())

df.columns = clean_cols
```

```python
df = df.drop(columns = ['index'])
```

```python
df
```

```python
suburbDf = df[['acara_sml_id','suburb','state','postcode','lga_code']]
```
